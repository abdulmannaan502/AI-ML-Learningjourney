# EDA + Visualization

```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("titanic.csv")
```

## 🧼 2. Clean the Data

```python
#Missing Values
df.dropna
```

```python
df.fillna(0)
```

```python
df['Age'].fillna(df['Age'].mean(), inplace = True)
```

```python
#Duplicate Rows
df.duplicated().sum()
df.drop_duplicates(inplace=True) 
```

```python
#Fixing Datatypes
df['Fare'] = df['Fare'].astype(float)
```

## 🔍 3. Discover Patterns

```python
df.select_dtypes(include=['number']).corr()
```

```python
#Survival rate by gender
df.groupby('Sex')['Survived'].mean()
```

```python
#Average Fare by Passenger Class
df.groupby('Pclass')['Fare'].mean()
```

```python
#Value counts
df['Age'].value_counts() # Count of each category
```

## 📊 4. Visualize the Data

```python
#Histogram
df['Age'].hist(bins=10)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
#Bar Chart
df['Sex'].value_counts().plot(kind='bar', color=['skyblue', 'pink'])
plt.title("Gender Count")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
#Pie Chart
df['Survived'].value_counts().plot.pie(autopct='%1.1f%%', labels=['No', 'Yes'])
plt.title("Survival Rate")
plt.ylabel("")  # Remove y-label
plt.show()
```

```python
#Line Plot
df.plot(x='Age', y='Survived', kind='line')
plt.title("Survived/Age")
plt.xlabel("Age")
plt.ylabel("Survived")
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['number'])
```

```python
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
